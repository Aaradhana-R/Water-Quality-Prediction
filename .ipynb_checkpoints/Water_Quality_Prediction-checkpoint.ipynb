## Three columns have missing values i.e ph, sulfate, and Trihalomethanes and all are numeric. So We try to fill the ph column , Trihalomethanes column and sulfate column by imputaion method

```python
plt.figure(figsize= (5,2))
sns.boxplot(x = 'ph', data = df)
plt.show()
```

```python
plt.figure(figsize= (5,2))
sns.boxplot(x='Sulfate', data = df)
plt.show()
```

```python
plt.figure(figsize= (5,2))
sns.boxplot(x='Trihalomethanes', data = df)
plt.show()
```

### Here all contains outliers so we fill the missing values through median

```python
# lets fill the missing value of ph , Trihalomethanes and sulfate columns through median imputation

median_value1 = df['ph'].median()
median_value2 = df['Sulfate'].median()
median_value3 = df['Trihalomethanes'].median()
```

```python
df['ph'].fillna(median_value1, inplace=True)
df['Sulfate'].fillna(median_value2, inplace=True)
df['Trihalomethanes'].fillna(median_value3, inplace=True)
```
